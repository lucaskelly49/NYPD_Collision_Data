```python
# 2013 was dropped because of it's significance as an outlier

year = [2014, 2015, 2016, 2017, 2018]
collisions = [206030, 217692, 229787, 230997, 231467]

collision_year = pd.DataFrame(
    {'year': year,
     'collisions': collisions,
    })
```

```python
collision_year
```

```python
X = collision_year['year']
Y = collision_year['collisions']
```

```python
plt.scatter(X, Y, alpha=0.5)
plt.title('Total NYC Vehicle Collisions Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Collisions')
plt.show()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

# build the formula 
f = 'collisions~year'
# create a fitted model in one line
model = smf.ols(formula=f, data=collision_year).fit()
```

```python
model.summary()
```
